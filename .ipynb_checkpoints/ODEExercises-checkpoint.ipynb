Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def Euler(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ dydx[i] * dx
    return x,y
x1,y1 = Euler(100,0.25)
x2,y2 = Euler(100,0.02)
ii = (x1==1)
jj = (x1==1)
print y1[ii]
print y2[jj]
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution
y=0.25e2x−0.5x−0.25
on one plot. Include a legend. Which step size gives a more accurate result?

```python
x1,y1 = Euler(100,0.25)
x2,y2 = Euler(100,0.02)
y = 0.25 * np.exp(2*x2) - 0.5*x2 - 0.25
plt.title("y vs x")
plt.plot(x2,y,'g-',label="exact")
plt.plot(x1,y1, 'r-', label="step size = 0.25")
plt.plot(x2,y2, 'b-', label="step size = 0.02")
plt.axis([0,12,0,100])
plt.xlabel(r'$x$',fontsize=20)
plt.ylabel(r'$y$',fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def EulerCromer(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ dydx[i+1] * dx
    return x,y
x1,y1 = EulerCromer(100,0.25)
x2,y2 = EulerCromer(100,0.02)
def MidPoint(xmax,dx):
    a = -9.81 #g in m/s^2
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ 0.5 * (dydx[i] + dydx[i+1])* dx
    return x,y
x3,y3 = MidPoint(100,0.25)
x4,y4 = MidPoint(100,0.02)
```

```python
x1,y1 = Euler(100,0.25)
x2,y2 = Euler(100,0.02)
x3,y3 = EulerCromer(100,0.25)
x4,y4 = EulerCromer(100,0.02)
x5,y5 = MidPoint(100,0.25)
x6,y6 = MidPoint(100,0.02)
exact1 = 0.25 * np.exp(2*x1) - 0.5*x1 - 0.25
exact2 = 0.25 * np.exp(2*x2) - 0.5*x2 - 0.25
percentError1 = 100*abs( y1 - exact1 )/exact1
percentError2 = 100*abs( y2 - exact2 )/exact2
percentError3 = 100*abs( y3 - exact1 )/exact1
percentError4 = 100*abs( y4 - exact2 )/exact2
percentError5 = 100*abs( y5 - exact1 )/exact1
percentError6 = 100*abs( y6 - exact2 )/exact2
plt.axis([0,10,1,100])
plt.semilogy(x1, percentError1, label='Euler 0.25')
plt.semilogy(x2, percentError2, label='Euler 0.02')
plt.semilogy(x3, percentError3, label='Euler-Cromer 0.25')
plt.semilogy(x4, percentError4, label='Euler-Cromer 0.02')
plt.semilogy(x5, percentError5, label='Midpoint 0.25')
plt.semilogy(x6, percentError6, label='Midpoint 0.02')
plt.ylabel("percent error with exact")
plt.legend(loc="lower right")
plt.show()
```

We can use numerical solvers and widgets to explore the dynamics of a projectile that is subject to a drag force proportional to the square of its velocity - "quadratic drag".

The force on the projectile is given by
∑F⃗ =ma⃗ =mg⃗ −cv2v^=mg⃗ −cvv⃗ =mg⃗ −cv2x+v2y‾‾‾‾‾‾‾√v⃗ 

Where the second term is the drag force, in the direction opposite that of the velocity vector. Re-writing these in terms of the first-order differential equations for vx and vy gives

mdvxdt=−cv2x+v2y‾‾‾‾‾‾‾√vx
mdvydt=mg−cv2x+v2y‾‾‾‾‾‾‾√vy

taking the positive y direction to be down. Notice that these are coupled equations. There are both vx and vy terms in each one. The drag coefficient c depends on the shape and density of the projectile. We can re-write this system of equations in terms of the terminal velocity - the point at which the gravitational force balances the drag force. This happens when mg=cv2. Let's denote the terminal velocity as vter=mg/c‾‾‾‾‾√ and re-write our equations in terms of vter.

dvxdt=−gmvxv2terv2x+v2y‾‾‾‾‾‾‾√
dvydt=−gm(1+vyv2terv2x+v2y‾‾‾‾‾‾‾√)

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def Projectile_deriv(y,t):
    ay = -9.81 #g in m/s^2
    vx0 = y[0]
    vy0 = y[1]
    vterm = y[2]
    dvxdt = (ay*vx0/vterm**2) * np.sqrt(vy0**2 + vx0**2)
    dvydt = ay * (1 + (vy0/vterm**2) * np.sqrt(vy0**2 + vx0**2))
    return np.array([dvxdt,dvydt])
```

```python
def Projectile(v0=30,th0=50*np.pi/180,time=200,vterm=44):
    t = np.arange(0.,time+0.001,0.001)
    vx0 = v0 * np.cos(th0)
    vy0 = v0 * np.sin(th0)
    x0 = 0
    y0 = 0
    init = np.array([vx0,vy0,vterm])
    result = odeint(Projectile_deriv,init,t)
    print result
    vx = result[:,0]
    vy = result[:,1]
    x = vx * t
    y = vy * t
    plt.plot(x,y,label="Quadratic Drag")
    plt.xlabel("x")
    plt.ylabel("y(x)")
    plt.legend()
    plt.axis([0,50,0,20])
    plt.show()
Projectile() 
```
